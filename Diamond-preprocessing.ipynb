```python
###capping method: z score-> std normal distribution

import scipy.stats as stats
data['new_x'] = stats.zscore(data['x'])
data[(data['new_x']>-3) & (data['new_x']<3)].index
data.drop(data[(data['new_x']<-3) | (data['new_x']>3)].index)
```

```python
### treating outliers using IQR
Q1=data['table'].quantile(0.25)
Q2=data['table'].quantile(0.5)
Q3=data['table'].quantile(0.75)
```

```python
IQR = Q3 - Q1
LC = Q1 - (1.5 * IQR)
UC = Q3 + (1.5 * IQR)
print(LC,UC)
```

```python
sns.distplot(data['table'])
plt.axvline(UC, color = "r")
plt.axvline(LC, color = "r")
plt.axvline(Q1, color = "g")
plt.axvline(Q3, color = "g")
plt.show()
```

```python
data[(data['table'] < LC) | (data['table'] > UC)].index
```

```python
data = data.drop(data[(data['table'] < LC) | (data['table'] > UC)].index) 
```
