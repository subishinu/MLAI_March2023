```python
from sklearn.preprocessing import scale
```

```python
X = pd.DataFrame(scale(data), index = data.index, columns = data.columns)
```

```python
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
```

```python
inertia_list = []

for i in range(1,26):
  km = KMeans(n_clusters = i, init = "k-means++")
  km.fit(X)
  inertia = km.inertia_
  inertia_list.append(inertia)
```

```python
plt.figure(figsize=(10,10))
plt.plot(range(1,26), inertia_list)
plt.xlabel('Number of clusters')
plt.ylabel('inertia_list')
plt.show()
```

```python
from sklearn.decomposition import PCA
```

```python
pc = PCA(n_components = 30)  
```

```python
pc_array = pc.fit_transform(X)
```

```python
pc_array.shape
```

```python
pc_data = pd.DataFrame(pc_array, columns = ['PCA%i' % i for i in range(30)])
```
